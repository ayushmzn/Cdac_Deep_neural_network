```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
start = -10
end = 10
samples = 100
x = np.linspace(start, end, samples)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
res_sig = sigmoid(x)
plt.plot(x, res_sig)
plt.grid()
plt.title('sigmoid function')
plt.xlabel('-----x----->')
plt.ylabel('-----s(x)----->')
plt.show()
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
res_tan = tanh(x)
plt.plot(x, res_tan)
plt.grid()
plt.title('tanh function')
plt.xlabel('-----x----->')
plt.ylabel('-----s(x)----->')
plt.show()
```
