```python
from keras.datasets import mnist
from matplotlib import pyplot
```

```python
(xtrain, ytrain), (xtest, ytest) = mnist.load_data()
```

```python
for i in range(0, 9):
    pyplot.subplot(330+1+i)
    pyplot.imshow(xtrain[i], cmap=pyplot.get_cmap('gray'))
pyplot.show()    
```

```python
xtrain = xtrain.reshape(xtrain.shape[0], 28, 28, 1)
xtest = xtest.reshape(xtest.shape[0], 28, 28, 1)
```

```python
xtrain = xtrain.astype('float32')
xtest = xtest.astype('float32')
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rotation_range=100)
train_datagen.fit(xtrain)
```

```python
for xbetch, ybetch in train_datagen.flow(xtrain, ytrain, batch_size=9):
    for i in range(0, 9):
        pyplot.subplot(330+1+i)
        pyplot.imshow(xbetch[i], cmap=pyplot.get_cmap('gray'))
    pyplot.show()
    break
```
